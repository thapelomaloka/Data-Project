# <span style = "color:#38D0F2">Import My Data File In the Format of CSV</span>

```python
# Read the CSV file into a DataFrame.
sales_data = pd.read_csv('./Online_Shop_Customer_Sales_Data.csv')
```

# <span style = "color:#38D0F2">Data Cleaning</span>

### <span style = "color:#38D0F2">Check for Missing Values</span>

```python
# Check for missing values
missing_values = sales_data.isna()

# Count the number of missing values in each column
num_missing = missing_values.sum()

# Print the number of missing values in each column
print(num_missing)
```

### <span style = "color:#38D0F2">For each row it appears that there are no missing vaues</span>
### <span style = "color:#38D0F2">Check for data types</span>


```python
sales_data.dtypes
```

## <span style = "color:#38D0F2">Replace 0 with 'Male' and 1 with 'Female' in the Gender column</span> 
## <span style = "color:#38D0F2">Replace 0 with Chrome, 1 with Safari, 2 with Edge, 3 with Other in the Browser column</span>
## <span style = "color:#38D0F2">Replace 0 with Digital Wallets, 1 with Card, 2 with PayPal, 3 with Other in the Pay Method column</span>
## <span style = "color:#38D0F2">Replace 0 with not subscribed, 1 with subscribed in the Newsletter column</span>
## <span style = "color:#38D0F2">Replace 0 with not used, 1 with used in the Voucher column</span>


```python
sales_data['Gender'] = sales_data['Gender'].replace({0: 'Male', 1: 'Female'})

# Replace values in the Browser column
sales_data['Browser'] = sales_data['Browser'].replace({0: 'Chrome', 1: 'Safari', 2: 'Edge', 3: 'Other'})

# Replace values in the Pay Method column
sales_data['Pay_Method'] = sales_data['Pay_Method'].replace({0: 'Digital Wallets', 1: 'Card', 2: 'PayPal', 3: 'Other'})

# Replace values in the Newsletter column
sales_data['Newsletter'] = sales_data['Newsletter'].replace({0: 'not subscribed', 1: 'subscribed'})

# Replace values in the Voucher column
sales_data['Voucher'] = sales_data['Voucher'].replace({0: 'not used', 1: 'used'})


sales_data.head()
```

## <span style = "color:#38D0F2"> Create a new Column with the Month only</span> 


```python
sales_data['Month'] = sales_data['Purchase_DATE'].str[3:5]

sales_data.head()
```

## <span style = "color:#38D0F2"> Convert the new Column Month into an Interger </span>


```python
sales_data['Month'].astype(int)
```

## <span style = "color:#38D0F2">Rename the the N_Purchases to Number_Purchases & Time_Spent to Time_Spent(Sec) Column</span> 


```python
sales_data.rename(columns = {'Time_Spent' : 'Time_Spent(Sec)', 'N_Purchases' : 'Number_Purchases'},inplace = True) 
sales_data.head()
```

```python
sales_data.to_csv('Sales_Data_Final.csv', index=False)
```

# <span style = "color:#38D0F2">Descriptive Statistics. </span>

## <span style = "color:#38D0F2">Age: calculate the mean, median, mode, minimum, maximum,standard deviation and Inter Quartile Rage of the age of customers.</span>

```python
#Sort the Age column from low to high
SortAge = sales_clean_data.sort_values(by="Age")
SortAge['Age'].describe()
```

```python
#The mode of the Age:
Mode = SortAge['Age'].mode()[0]
print('Mode:', Mode)

#The Inter Quartile Range of the Age:
Q1 = SortAge['Age'].quantile(0.25)
Q3 = SortAge['Age'].quantile(0.75)

IQR = Q3 - Q1

print('IQR:', IQR)
```

### <span style = "color:#38D0F2"> Import Numpy to allow to round Up/down Numbers.</span>

```python
import numpy as np
```

## <span style = "color:#38D0F2"> Gender: count the number of customers for each gender, calculate the percentage of male and female customers (Distribution).</span>

```python
CountMale = 0
CountFemale = 0

for i in sales_clean_data['Gender']:
    if(i == 'Male'):
        CountMale = CountMale +1
    else:
        CountFemale = CountFemale +1
        
print('The total number of Males is: ', CountMale)
print('The total number of Females is: ', CountFemale)

#Calculate the percentage of Male and Female customers
total_no_cus = CountFemale + CountMale
percentage_males = np.ceil((CountMale / total_no_cus) * 100)
percentage_females = np.floor((CountFemale / total_no_cus) * 100)

print('The percentage of Female customers is: ',percentage_females,"%")
print('The percentage of Male customers is: ',percentage_males,"%")
```

### <span style = "color:#38D0F2">Pie Chart to Show distribution of Male and Females.</span>

### <span style = "color:#38D0F2">Import Seaborn and Matplotlib for my Pie Chart.</span>

```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# create data
labels = ['Female', 'Male']
sizes = [67.0, 33.0]

# Create a pie chart
sns.set_style("whitegrid")
plt.figure(figsize=(6,6))
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['#DA0037', '#FF4301'])
plt.title('Percentage of Female and Male Customers')
plt.show()
```
